```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns to explore
categorical_columns = ["Country Code", "City", "Cuisines"]

# Loop through each categorical column and plot distribution
for col in categorical_columns:
    plt.figure(figsize=(12, 5))

    # Plot using seaborn countplot
    sns.countplot(y=df[col], order=df[col].value_counts().index[:10], palette="viridis")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.title(f"Distribution of {col} (Top 10)")

    # Show the plot
    plt.show()
```

```python
import plotly.express as px

# List of categorical columns to explore
categorical_columns = ["Country Code", "City", "Cuisines"]

# Loop through each categorical column and plot distribution
for col in categorical_columns:
    top_values = df[col].value_counts().nlargest(10)  # Get top 10 most frequent categories

    # Create interactive bar chart using Plotly
    fig = px.bar(
        x=top_values.values,
        y=top_values.index,
        orientation='h',
        text=top_values.values,  # Display values on hover
        title=f"Distribution of {col} (Top 10)",
        labels={'x': 'Count', 'y': col},
        color=top_values.values,  # Color based on frequency
        color_continuous_scale='viridis'
    )

    fig.update_traces(texttemplate='%{text}', textposition='outside', hoverinfo='x+y')
    fig.show()
```

```python
import plotly.express as px

# Ensure latitude & longitude columns exist
if "Latitude" in df.columns and "Longitude" in df.columns:
    fig = px.scatter_mapbox(
        df,
        lat="Latitude",
        lon="Longitude",
        hover_name="City",  # Display city name on hover
        hover_data=["Cuisines"],  # Show additional info
        zoom=2,  # Initial zoom level
        height=600,
        title="Restaurant Locations on Map"
    )

    fig.update_layout(
        mapbox_style="open-street-map",  # Use open-source tiles
        margin={"r":0, "t":50, "l":0, "b":0}
    )

    fig.show()
else:
    print("Latitude and Longitude columns are missing in the dataset.")
```

```python
import plotly.express as px

# Top 10 Cities with Most Restaurants
top_cities = df["City"].value_counts().nlargest(10)

fig_city = px.bar(
    x=top_cities.index,
    y=top_cities.values,
    text=top_cities.values,
    title="Top 10 Cities with Most Restaurants",
    labels={'x': 'City', 'y': 'Number of Restaurants'},
    color=top_cities.values,
    color_continuous_scale='viridis'
)
fig_city.update_traces(texttemplate='%{text}', textposition='outside')
fig_city.show()

# Top 10 Countries with Most Restaurants
top_countries = df["Country Code"].value_counts().nlargest(10)

fig_country = px.bar(
    x=top_countries.index,
    y=top_countries.values,
    text=top_countries.values,
    title="Top 10 Countries with Most Restaurants",
    labels={'x': 'Country Code', 'y': 'Number of Restaurants'},
    color=top_countries.values,
    color_continuous_scale='magma'
)
fig_country.update_traces(texttemplate='%{text}', textposition='outside')
fig_country.show()
```

```python
# Average Rating by City
avg_rating_city = df.groupby("City")["Aggregate rating"].mean().reset_index()
avg_rating_city = avg_rating_city.sort_values(by="Aggregate rating", ascending=False).head(10)

# Plot the correlation between City and Ratings
fig_city_rating = px.scatter(
    avg_rating_city,
    x="City",
    y="Aggregate rating",
    size="Aggregate rating",
    title="Top 10 Cities with Highest Average Restaurant Ratings",
    labels={'Aggregate rating': 'Average Rating', 'City': 'City'},
    color="Aggregate rating",
    color_continuous_scale="plasma"
)
fig_city_rating.show()

# Average Rating by Country
avg_rating_country = df.groupby("Country Code")["Aggregate rating"].mean().reset_index()
avg_rating_country = avg_rating_country.sort_values(by="Aggregate rating", ascending=False)

# Plot the correlation between Country and Ratings
fig_country_rating = px.scatter(
    avg_rating_country,
    x="Country Code",
    y="Aggregate rating",
    size="Aggregate rating",
    title="Average Restaurant Ratings by Country",
    labels={'Aggregate rating': 'Average Rating', 'Country Code': 'Country'},
    color="Aggregate rating",
    color_continuous_scale="cividis"
)
fig_country_rating.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Group by price range and calculate average rating
avg_rating_per_price = df.groupby("Price range")["Aggregate rating"].mean().reset_index()

# Merge with the dataset to get corresponding rating colors
rating_color_map = df[["Price range", "Rating color"]].drop_duplicates()
avg_rating_per_price = avg_rating_per_price.merge(rating_color_map, on="Price range", how="left")

# Define a dictionary to map Rating Colors to valid HEX codes
color_mapping = {
    "Dark Green": "#006400",  # Dark Green
    "Green": "#008000",       # Green
    "Yellow": "#FFFF00",      # Yellow
    "Orange": "#FFA500",      # Orange
    "White": "#D3D3D3",       # Light Gray (White is not visible, so using Gray)
    "Red": "#FF0000"          # Red
}

# Apply color mapping, filling NaN values with a default color
avg_rating_per_price["Mapped_Color"] = avg_rating_per_price["Rating color"].map(color_mapping)
avg_rating_per_price["Mapped_Color"].fillna("#D3D3D3", inplace=True)  # Default color (light gray) for NaNs

# Plot barplot with mapped colors
plt.figure(figsize=(8, 5))
sns.barplot(x="Price range", y="Aggregate rating", data=avg_rating_per_price, 
            palette=avg_rating_per_price["Mapped_Color"].values, edgecolor="black")

# Add labels and title
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.title("Average Rating by Price Range with Rating Colors")

# Show the plot
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
